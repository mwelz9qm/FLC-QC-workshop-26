## Part 2 — Meet your quantum circuit

The cell below builds a quantum circuit with **1 qubit** and **1 classical bit** (a normal bit to store the answer when we measure). Then it draws it.

Run it and look at the picture.

````python
qc = QuantumCircuit(1, 1)   # 1 qubit, 1 classical bit for the result

# ── the middle of the circuit is... empty! ──

qc.measure(0, 0)            # look at qubit 0, store the answer in bit 0

qc.draw("mpl")              # draw the circuit
````

That black dial-looking box is the **measurement** — the "hand-slam" from the penny game. But right now nothing happens before it: the qubit starts as a boring, definite 0 and we just... look at it.

Let's confirm this circuit is boring. **🔮 PREDICT FIRST:** if we run it 1000 times, what will the histogram look like?

````python
result = simulator.run(qc, shots=1000).result()   # shots = how many times to repeat
counts = result.get_counts()
plot_histogram(counts)
````

One bar. 1000 out of 1000 zeros. The universe's most boring circuit. 😴

## Part 4 — Rebuild the Penny Flip Duel (and prove Q can't lose)

The game from today: penny starts heads-up (that's `0`), moves go **Q, P, Q**.
- Q's quantum move = `h`
- P's flip = `x` (the NOT gate — the plain old flip)

The cell below plays the version where **P flips**. Run it.

````python
duel = QuantumCircuit(1, 1)
duel.h(0)      # Q's move: spin it
duel.x(0)      # P's move: flip! (take that, Q)
duel.h(0)      # Q's move: un-spin it
duel.measure(0, 0)

counts = simulator.run(duel, shots=1000).result().get_counts()
print(counts, " — remember: 0 = heads = Q wins")
plot_histogram(counts)
````

**🔮 PREDICT FIRST:** will removing P's flip change the outcome?

<details><summary>👉 What you should find</summary>

You have now reproduced a real research result: D. Meyer, *Quantum Strategies* (1999) — one of the founding papers of quantum game theory. Day one. Not bad.

<details><summary><b>E1 — The rigged coin</b></summary>

```python
e1.x(0)
```
`x` is the NOT gate: definite 0 → definite 1. No randomness at all.

<details><summary><b>E2 — Order matters</b></summary>

With the extra `h` appended: A (`x,h,h`) collapses to just `x` → **100% ones**, while B (`h,x,h`) is the penny-flip duel → **100% zeros**. Opposite answers! So the two original 50/50 states really were different — the difference (a relative minus sign, called *phase*) is invisible until another gate makes it visible. Interference between hidden phases is the engine behind every algorithm in Session 4.
</details>

<details><summary><b>E3 — The weighted coin</b></summary>

The probability of measuring 0 is `cos²(angle/2)`, so you need `angle = 2·arcsin(√0.15) ≈ 0.795` radians. Anything from ~0.75–0.85 looks right at 1000 shots.
</details>

<details><summary><b>E4 — Two coins</b></summary>

Four bars — `00, 01, 10, 11` — each ≈250. Two independent spinning coins = a superposition of all four two-bit strings. Each added qubit **doubles** the number of outcomes; that exponential blow-up is where quantum computers get their room to work.
</details>

<details><summary><b>E5 — Uniform randomizer</b></summary>

```python
def uniform_randomizer(n):
    qc = QuantumCircuit(n, n)
    for i in range(n):
        qc.h(i)
    qc.measure(range(n), range(n))
    return qc
```
